Podemos pasar otras no sÃ³lo una funcion, ademÃ¡s de funciones definidas con `aggregate`

```python
def mean_kilo(x):
    return np.mean(x)/1000
```

```python
df.groupby(['cut', 'color'])['price'].aggregate(['min', np.mean, max])
```

```python
df.groupby(['cut', 'color'])['price'].aggregate(['min', np.mean, mean_kilo])
```

```python
dict_agg={'carat':[min,max], 'price':[np.mean,mean_kilo]}
#diccionario de instrucciones
#varios calculos estadÃ­sticos para diferentes categorÃ­as, 
```

```python
df.groupby(['cut', 'color']).aggregate(dict_agg)
```

Agrupa primero por corte, y luego agrupa por color ya para cada grupo obtiene los calculos indicados en aggregate, min y mac de kilates y medias de precio

AsÃ­ tenemos mÃºltiples Ã­ndices en columnas y filas


### filtros

```python
def f_filter(x):
    return mean_kilo(x['price'])>4
```

```python
df.groupby('cut').filter(f_filter)
```

```python
df.groupby('cut').filter(f_filter)['cut'].unique()
```
